### **Cek Outliers**

Pada tahap ini, kita akan mendeteksi adanya outliers pada beberapa kolom numerik menggunakan visualisasi boxplot.
Berdasarkan analisis sebelumnya, kolom yang memiliki persentase outliers adalah:

* Total Revenue (8.6%)
* Total Cost (10.7%)
* Total Profit (5.5%)

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

### **Outliers Values**

Mengecheck persetase outliers pada setiap kolom.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Terdapat 3 outliers pada kolom Total Revenue, Total Cost, dan Total Profit ditangani dengan metode clipping, yaitu membatasi nilai yang terlalu tinggi atau terlalu rendah agar tetap berada dalam batas wajar.

```python
columns_to_impute = ["Total Revenue", "Total Cost", "Total Profit"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

Pengecekan ulang dilakukan untuk memastikan bahwa proses penanganan outliers berhasil dan tidak ada lagi nilai ekstrem yang bisa mengganggu hasil analisis selanjutnya.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## **Construct Data**

Menambahkan kolom Profit Category dengan tujuan mengelompokkan transaksi berdasarkan besar kecilnya keuntungan dengan manfaat memudahkan melihat transaksi mana yang paling menguntungkan tanpa harus melihat angka satu per satu.

```python
def categorize_profit(profit):
    if profit < df['Total Profit'].quantile(0.33):
        return 'Low Profit'
    elif profit < df['Total Profit'].quantile(0.66):
        return 'Medium Profit'
    else:
        return 'High Profit'

df['Profit Category'] = df['Total Profit'].apply(categorize_profit)
```

Dari grafik terlihat bahwa jumlah unit yang terjual tersebar di berbagai rentang nilai, sehingga menunjukkan bahwa pembelian pelanggan cukup bervariasi dan tidak terpusat pada satu jumlah unit tertentu. Hal ini mengindikasikan bahwa pelanggan melakukan pembelian dengan kuantitas yang beragam pada setiap order.

### **Heatmap**

Pada visualisasi ini saya menggunakan kolom Units Sold, Total Revenue, dan Total Profit untuk menganalisis hubungan atau korelasi antar variabel yang berkaitan dengan performa penjualan. Kolom-kolom tersebut dipilih karena dapat menunjukkan bagaimana jumlah produk yang terjual mempengaruhi pendapatan serta keuntungan yang diperoleh perusahaan.

```python
plt.figure(figsize=(8, 6))

sns.heatmap(data=df[['Units Sold', 'Total Revenue', 'Total Profit']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')

plt.title('Korelasi antara Units Sold, Total Revenue, dan Total Profit')
plt.show()
```

Dari hasil visualisasi terlihat bahwa Total Revenue dan Total Profit memiliki korelasi yang sangat kuat (0.89), yang berarti semakin tinggi pendapatan maka keuntungan juga cenderung meningkat. Sementara itu, Units Sold memiliki korelasi sedang dengan Total Revenue (0.54) dan Total Profit (0.60), yang menunjukkan bahwa jumlah unit yang terjual juga berpengaruh terhadap pendapatan dan keuntungan, meskipun tidak sekuat hubungan antara revenue dan profit.
